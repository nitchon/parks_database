```python
import pandas as pd
import requests
import datetime as dt
import numpy as np
```

```python
# Load data
National_Parks_cleaned_df = pd.read_csv('National_Parks_cleaned.csv')
National_Parks_cleaned_df

```

```python
all_trails_filtered_df = pd.read_csv('alltrails_filtered.csv')
all_trails_filtered_df
```

```python
exploratory_df = National_Parks_cleaned_df
exploratory_df.describe()
```

```python
exploratory_df2 = all_trails_filtered_df
exploratory_df2.describe()
```

```python
all_trails_filtered_df2 = all_trails_filtered_df.groupby(['park_code']).count()['trail_id']
all_trails_filtered_df2
```

```python
all_trails_filtered_df3 = all_trails_filtered_df.groupby(['park_code']).mean()['difficulty_rating']
all_trails_filtered_df3
```

```python
all_trails_filtered_df4 = all_trails_filtered_df.groupby(['park_code']).mean()['avg_rating']
all_trails_filtered_df4
```

```python
summary_table = pd.DataFrame({'num_of_trails':all_trails_filtered_df.groupby('park_code').count()['trail_id'],
                'avg_difficulty':all_trails_filtered_df.groupby('park_code').mean()['difficulty_rating'],
                'avg_rating':all_trails_filtered_df.groupby('park_code').mean()['avg_rating']})

summary_table = summary_table.reset_index(drop=False)
summary_table
```

```python
Combined_Data = summary_table.merge(National_Parks_cleaned_df, how = 'outer', on='park_code')
Combined_Data
```

```python
Combined_Data = Combined_Data.fillna(0)
Combined_Data
```

```python
## convert to csv 
Combined_Data.to_csv('national_parks_cleaned_with_avgs.csv',index=False)
```
