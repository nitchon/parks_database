```python
import pandas as pd
import requests
import datetime as dt
import numpy as np
```

```python
# Load the National Parks data into park_df
park_df = pd.read_csv('Resources/National_Parks.csv')
park_df
```

```python
# Rename the columns to names without blanks 
park_df.rename(columns= {"parkCode": "park_code", "Park Name": "park_name", "Fee per Car":"fee_per_car",
                        "Fee per Person":"fee_per_person","Street Address":"street_address","City":"city",
                        "State": "state","Zip Code":"zip","Phone Number":"phone","Email Address":"email",
                        "Lat":"lat","Lng":"lng","Weather Summary":"weather_summary"
                        }, inplace = True)
park_df
```

```python
#Save it to National Parks Cleaned CSV file
park_df.to_csv('Resources/National_Parks_cleaned.csv',index=False)
```

```python
# Get the trail data into a dataframe
alltrails_df = pd.read_csv('Resources/all_trails_cleaned.csv')

# Rename the column with trail name
alltrails_df.rename(columns= {"name": "trail"}, inplace = True)

# Drop the unwanted columns
alltrails_df = alltrails_df.drop(['area_name','city_name','state_name','country_name','visitor_usage'], axis = 1)

# # Make the Park name it's index
alltrails_df.set_index('park_code')

alltrails_df
```

```python
alltrails_df = alltrails_df.drop(alltrails_df.iloc[:, 0:1],
                       axis = 1)
alltrails_df
```

```python
# Filtering out the trails that have an average rating of less than 2, numer of reviews lesser than 10, and 
# sorting by popularity
alltrails_filtered = alltrails_df[alltrails_df['avg_rating'] >= 2]
alltrails_filtered = alltrails_filtered[alltrails_filtered['num_reviews'] >= 10]
alltrails_filtered.sort_values(by='popularity', inplace=True, ascending=False)

alltrails_filtered
```

```python
# Save all trail dataframe to a csv file in the Resources folder
alltrails_filtered.to_csv('Resources/alltrails_filtered.csv',index=False)
```
