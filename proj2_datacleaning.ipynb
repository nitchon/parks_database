```python
import pandas as pd
import requests
import datetime as dt
import numpy as np
```

```python
# Load the National Parks data into park_df
park_df = pd.read_csv('Resources/National_Parks.csv')
park_df
```

```python
# Rename the columns to names without blanks 
park_df.rename(columns= {"parkCode": "park_code", "Park Name": "park_name", "Fee per Car":"fee_per_car",
                        "Fee per Person":"fee_per_person","Street Address":"street_address","City":"city",
                        "State": "state","Zip Code":"zip","Phone Number":"phone","Email Address":"email",
                        "Lat":"lat","Lng":"lng","Weather Summary":"weather_summary"
                        }, inplace = True)
park_df
```

```python
#Save it to National Parks Cleaned CSV file
park_df.to_csv('Resources/National_Parks_cleaned.csv',index=False)
```

```python
# Get the trail data into a dataframe
alltrails_df = pd.read_csv('Resources/AllTrails data - nationalpark.csv')

# Rename the columns
alltrails_df.rename(columns= {"name": "trail", "area_name": "park_name"}, inplace = True)

# Drop the unwanted columns
alltrails_df.drop(['city_name','state_name','country_name','visitor_usage'], axis = 1)

# Change the index to park name
alltrails_df.set_index('park_name')

alltrails_df
```

```python
# From the National parks dataframe, copy the park name and park code into another dataframe - park_code_df
park_code_df = park_df[["park_name","park_code"]].copy()
park_code_df

# Make the Park name it's index
park_code_df.set_index('park_name')
```

```python
# Add the park code to the all trail dataframe
trails_park_code_df = alltrails_df.merge(park_code_df, how='left')

trails_park_code_df
```

```python
# Save all trail dataframe to a csv file in the Resources folder
park_df.to_csv('Resources/trails_park_code_cleaned.csv',index=False)
```
